```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'blocking_shift_scores.csv'
data = pd.read_csv(file_path, header=None)
data.columns = ['message_size', 'bandwidth', 'latency']
data.head()
```

```python
# Plotting the bandwidth vs. message size
plt.figure(figsize=(8, 5))
plt.plot(data['message_size'], data['bandwidth'], marker='o', linestyle='-', color='b')
plt.title('Bandwidth vs. Message Size')
plt.xlabel('Message Size (bytes)')
plt.ylabel('Bandwidth (MB/s)')
plt.grid(True)
plt.xscale('log')  # Using a logarithmic scale for the x-axis to better visualize the data
plt.yscale('linear')
plt.savefig('report/part3_1.png')
plt.show()
```

```python
# Plotting the latency vs. message size
plt.figure(figsize=(8, 5))
plt.plot(data['message_size'], data['latency'], marker='o', linestyle='-', color='b')
plt.title('Latency vs. Message Size')
plt.xlabel('Message Size (bytes)')
plt.ylabel('Latency (us)')
plt.grid(True)
plt.xscale('log')  # Using a logarithmic scale for the x-axis to better visualize the data
plt.yscale('linear')
plt.savefig('report/part3_2.png')
plt.show()
```
